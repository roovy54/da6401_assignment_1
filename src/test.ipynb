```python
import numpy as np

# Load .npz (weights stored as arrays)
data = np.load("best_model.npz", allow_pickle=True)

# If saved as dictionary inside npz
try:
    model_dict = data["arr_0"].item()  # common case if dict was saved
except:
    model_dict = {key: data[key] for key in data.files}

# Save as .npy (pickle-based dictionary)
np.save("best_model.npy", model_dict, allow_pickle=True)

print("Converted best_model.npz -> best_model.npy")
```
